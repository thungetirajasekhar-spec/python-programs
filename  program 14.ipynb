```python
import numpy as np
import pandas as pd
data={
    'NAME':['RAJU','VAMSI','RAVI','PRASAD'],
    'DEPARTMENT':['AIDS','CSE','CS','EEE'],
    'FEES':[20000,30000,10050,np.nan],
    'age':[20,19,30,np.nan]
}
df=pd.DataFrame(data)
df_renamed = df.rename(columns={'NAME':'N','DEPARTMENT':'D','FEES':'F'})
df['FEES'].fillna('50000',inplace=True)
df['FEES']=df['FEES'].astype(int)
df['age'].fillna('19',inplace=True)
df['age']=df['age'].astype(int)
print('data of the students:')
print(df)
```

```python
import pandas as pd
data = {
 'Color': ['Red', 'Blue', 'Green', 'Blue', 'Red'],
 'Size': ['S', 'M', 'L', 'XL', 'L']
}
df = pd.DataFrame(data)
# Create dummy variables
df_dummies = pd.get_dummies(df, columns=['Color', 'Size'])
print("Original DataFrame:\n", df)
print("\nDataFrame with Dummy Variables:\n", df_dummies)
```

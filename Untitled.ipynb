```python
import numpy as np
import matplotlib.pyplot as plt

# Years to plot
years = np.array([2023, 2024, 2025])

# --- Example data (replace with your real numbers) ---
# These could be production index, acreage (in 000s), % adoption, etc.
modern = np.array([100.0, 108.8, 118.5])   # modern farming (example: index / units)
organic = np.array([20.0, 30.0, 45.0])     # organic farming (example: index / units)

# Create the plot
plt.figure(figsize=(9,5))
plt.plot(years, modern, marker='o', linewidth=2, label='Modern farming')
plt.plot(years, organic, marker='o', linewidth=2, label='Organic farming (ZBNF)')

# Labels and title
plt.xlabel('Year')
plt.ylabel('Value (replace with your units)')
plt.title('Comparison: Modern farming vs Organic farming (2023-2025)')

# Grid, ticks and legend
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
plt.xticks(years)
plt.legend()

# Annotate exact values on the chart
for x, y in zip(years, modern):
    plt.annotate(f'{y:.1f}', xy=(x, y), xytext=(0, 6), textcoords='offset points', ha='center', fontsize=9)
for x, y in zip(years, organic):
    plt.annotate(f'{y:.1f}', xy=(x, y), xytext=(0, -10), textcoords='offset points', ha='center', fontsize=9)

# Save and show
plt.tight_layout()
plt.savefig('modern_vs_organic_2023_2025.png', dpi=150)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("emails.csv")
sns.countplot(x=df['spam'])
plt.title("Spam vs Ham Distribution")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("emails.csv")
plt.fill_between(df["the"], df["to"], color="green", alpha=0.5)
plt.xlabel("the")
plt.ylabel("to")
plt.title("Area Plot of 'to' Over 'the'")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("emails.csv")
x_column = "the"
y_column = "to"
plt.scatter(df[x_column], df[y_column], color="blue", alpha=0.7, edgecolors="black")
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.title(f"Scatter Plot of {x_column} vs {y_column}")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("emails.csv")
column_name = "the"
df[column_name].hist(bins=20, edgecolor="black")
plt.xlabel(column_name)
plt.ylabel("Frequency")
plt.title(f"Histogram of {column_name}")
plt.show()
```

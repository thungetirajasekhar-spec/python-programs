```python
import pandas as pd
import numpy as np
 
data = {
 'Age': [25, 30, 35, 30, 45],
 'Salary': [50000, 60000, 70000, 80000, 60000]
}
df = pd.DataFrame(data)
 
percentiles = [25, 50, 75]
 
age_percentiles = np.percentile(df['Age'], percentiles)
 
salary_percentiles = np.percentile(df['Salary'], percentiles)
print("Percentiles for Age:", age_percentiles)
print("Percentiles for Salary:", salary_percentiles)
```

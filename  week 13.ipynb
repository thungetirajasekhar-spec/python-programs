```python
import numpy as np
import matplotlib.pyplot as plt
uniform_sample=np.random.uniform(low=0.0,high=0.1,size=1000)
plt.hist(uniform_sample,bins=30,edgecolor='k')
plt.title('uniform distubution')
plt.xlabel('value')
plt.ylabel('value')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
mean=0
std_dev=1
normal_sample=np.random.normal(loc=mean,scale=std_dev,size=1000)
plt.hist(normal_sample,bins=30,edgecolor='k',density=True)
plt.title('normal distrubution')
plt.xlabel('value')
plt.ylabel('frequnency')
xmin,xmax=plt.xlim()
x=np.linspace(xmin,xmax,100)
p=np.exp(-0.5*((x-mean)/std_dev)**2)/(std_dev*np.sqrt(2*np.pi))
plt.plot(x,p,'k',linewidth=2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma
shape=2.0
scale=2.0
gamma_sample=gamma.rvs(a=shape,scale=scale,size=1000)
plt.hist(gamma_sample,bins=30,edgecolor='k',density=True)
plt.title('gamma distrubution')
plt.xlabel('value')
plt.ylabel('frequncy')
x=np.linspace(0,np.max(gamma_sample),1000)
plt.plot(x,gamma.pdf(x,a=shape,scale=scale),'r',lw=2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
scale=1.0
exponential_sample=np.random.exponential(scale=scale,size=1000)
plt.hist(gamma_sample,bins=30,edgecolor='k',density=True)
plt.hist('exponetial distubion')
plt.xlabel('value')
plt.ylabel('frequncy')
x=np.linspace(0,np.max(exponential_sample),1000)
plt.plot(x,(1/scale)*np.exp(-x/scale),'r',lw=2)
plt.show()
```

```python
def fibonacci_memo(n, memo={0: 0, 1: 1}):
    if n in memo:
        return memo[n]
    memo[n] = fibonacci_memo(n - 1, memo) + fibonacci_memo(n - 2, memo)
    return memo[n]

n = 100000000
fib_series = [fibonacci_memo(i) for i in range(n)]
print(fib_series)
```

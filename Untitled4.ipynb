```python
import pandas as pd
data={
    'NAME':['alice','bob','alice','charlie','bob','eve','bob'],
    'AGE':[25,30,25,35,30,40,54]
}
df=pd.DataFrame(data)
print("original data:")
print(df)
df_deduplicated=df.drop_duplicates()
print("\nDeduplicated data:")
print(df_deduplicated)
```

```python
import pandas as pd
data={
    'name':['alice','bob','alice','charlie','bob','eve','bob'],
    'age':[25,30,25,35,30,40,54]
}
df=pd.DataFrame(data)
print("original data:")
print(df)
df.replace({'alice':'alice'},inplace=True)
print("\nupdated data after replacing'alice'with'alice':")
print(df)
df.loc[df['age']==30,'age']=29
print("\nupdated data after replacing age 30 with 29:")
print(df)
```

```python
import pandas as pd
import numpy as np
data = {
 'A': [1, 2, np.nan, 4],
 'B': [np.nan, 2, 3, 4]
}
df = pd.DataFrame(data)
print("Original data:")
print(df)
df_filled = df.fillna(0)
print("\nData after replacing NaN with 0:")
print(df_filled)
df_filled_mean = df.fillna(df.mean())
print("\nData after replacing NaN with the column mean:")
print(df_filled_mean)
df_dropped = df.dropna()
print("\nData after dropping rows with NaN values:")
print(df_dropped)
 
```

```python
import pandas as pd
data={
    'name':['raj','ravi','vamsi','prasanth'],
    'age':[19,20,21,23]
}
df=pd.DataFrame(data)
print(df)
df = pd.DataFrame(data)
# Forward Fill
df_ffill = df.fillna(method='ffill')
# Backward Fill
df_bfill = df.fillna(method='bfill')
print("Original DataFrame:\n", df)
print("\nForward Fill DataFrame:\n", df_ffill)
print("\nBackward Fill DataFrame:\n", df_bfill)
```

```python
import pandas as pd
import numpy as np
data = {
 'A': [1, 2, np.nan, 4, 5],
 'B': [np.nan, 2, 3, 4, np.nan],
 'C': [1, np.nan, np.nan, 4, 5]
}
df = pd.DataFrame(data) 
df_filled_index = df.apply(lambda x: x.fillna(x.index))
print("Original DataFrame:\n", df)
print("\nDataFrame filled with index values:\n", df_filled_index)
```

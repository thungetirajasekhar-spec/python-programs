```python
import matplotlib.pyplot as plt
import numpy as np
# Data
labels =['A', 'B', 'C', 'D']
values = [4, 7, 1, 5]
# Prepare data
values += values[:1]  # Duplicate the first value at the end to close the circular plot
angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
angles += angles[:1]  # Duplicate the first angle at the end to close the circular plot
# Plot
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))  # Create a polar subplot
ax.fill(angles, values, color='purple', alpha=0.25)  # Fill the area inside the plot
ax.plot(angles, values, color='purple', linewidth=2)  # Plot the line
# Customization
ax.set_yticklabels([])  # Remove the radial (circular) grid labels
ax.set_xticks(angles[:-1])  # Set the angular ticks
ax.set_xticklabels(labels)  # Set the labels for the angular ticks
plt.show()  
```

```python
import matplotlib.pyplot as plt

 
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [25, 30, 15, 30]   

 
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
explode = (0.1, 0, 0, 0)   

 
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, colors=colors, explode=explode,
        autopct='%1.1f%%', startangle=140, shadow=True)

plt.title('Fruit Distribution')
plt.axis('equal')  
plt.show()
```

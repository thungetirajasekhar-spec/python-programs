```python
import pandas as pd
data = {
 'Company': ['A', 'A', 'A', 'B', 'B', 'B'],
 'Department': ['HR', 'IT', 'Finance', 'HR', 'IT', 'Finance'],
 'Employee': ['John', 'Alice', 'Steve', 'Bob', 'James', 'Linda'],
 'Salary': [50000, 70000, 60000, 55000, 65000, 62000]
}
df = pd.DataFrame(data)
print(df)
df.set_index(['Company', 'Department', 'Employee'], inplace=True)
print("\nDataFrame with Hierarchical Index:")
print(df)
unstacked = df.unstack(level='Department')
print(unstacked)
```

```python
import pandas as pd
data = {
 'Company': ['A', 'A', 'A', 'B', 'B', 'B'],
 'Department': ['HR', 'IT', 'Finance', 'HR', 'IT', 'Finance'],
 'Employee': ['John', 'Alice', 'Steve', 'Bob', 'James', 'Linda'],
 'Salary': [50000, 70000, 60000, 55000, 65000, 62000]
}
df = pd.DataFrame(data)
df.set_index(['Company', 'Department', 'Employee'], inplace=True)
print("\nDataFrame with Hierarchical Index:")
print(df)
average_salary = df.groupby(level=['Company', 'Department']).mean()
print("\nAverage Salary by Company and Department:")
print(average_salary)
```
